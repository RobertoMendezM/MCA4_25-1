```python

import numpy as np
import matplotlib.pyplot as plt


k = 1

# Definicnde la Derivada
def f(x,y):
    return -x**2 -(x -1)*y


#nx, ny = .3, .3
nx, ny = .2, .2
x = np.arange(-3, 3, nx)
y = np.arange(-2, 2, ny)

# MESHGRID
X, Y = np.meshgrid(x, y)

# Derivative
dy = f(X,Y)
dx =np.ones(X.shape)

# Normalizar
dyu = dy/np.sqrt(dx**2 + dy**2)
dxu = dx/np.sqrt(dx**2 + dy**2)

# Grficas
plt.quiver(X,Y,dxu,dyu, color = "orange",  headwidth = 2)
plt.xticks(x, rotation = 60)
plt.yticks(y)
plt.title("Directional Field y' = -x**2 -(x -1)*y", color='blue', fontsize ='x-large')
plt.show()
```
